```python
from matplotlib import ticker
formatter = ticker.ScalarFormatter(useMathText=True)
formatter.set_scientific(True)
```

```python
fig, axes = plt.subplots(2, 3, figsize=(20, 7), sharex=True)
env_names = ['hopper-medium-replay-v0', 'hopper-random-v0', 'walker2d-medium-v0',
            'walker2d-medium-expert-v0', 'halfcheetah-medium-expert-v0', 'halfcheetah-medium-v0']
for i, env_name in enumerate(env_names):
    x = i // 3
    y = i % 3
    ax = axes[x, y]
    data = get_all_datasets([f'data/ablation_results_finished/env={env_name}_reg='])
    plot_data(data, xaxis='GradientSteps', value='Performance', smooth=20, ax=ax)
    ax.set_title(env_name)
    ax.get_legend().remove()
#     ax.xaxis.set_major_formatter(formatter)

handles, labels = ax.get_legend_handles_labels()
fig.legend(handles, ['reg=kl, gp=false', 'reg=kl, gp=true', 'reg=mmd, gp=false', 'reg=mmd, gp=true'], 
           loc='lower center', ncol=4, bbox_to_anchor=(0.4, -0.1))
plt.savefig('/home/chi/Desktop/iclr21/figures/ablation_performance.pdf', bbox_inches='tight')
```
